```python
s1=input()
s2=input()
def Levenstein_dist(s1,s2):
    r,c=len(s1),len(s2)
    dp=[[-1 for j in range(0,c+1)]for i in range(0,r+1)]
    for i in range(0,r+1):
      dp[i][0]=i
    for j in range(0,c+1):
      dp[0][j]=j
    for i in range(1,r+1):
      for j in range(1,c+1):
        if(s1[i-1]==s2[j-1]):
            dp[i][j]=dp[i-1][j-1]
        else:
            dp[i][j]=1+min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])
    return dp[r][c]
print(Levenstein_dist(s1,s2))
```

Levenstein/edit distance function

Displaying our vocabulary from the Information retrieval system

```python
def k_nearest_word(s1,vocabulary,k):
    k_nearest_word=[]
    word_dist={}
    for word in vocabulary:
        word_dist[word]=Levenstein_dist(s1,word)
    word_dist=dict(sorted(word_dist.items(), key=lambda item: item[1]))
    ans=[]
    li=list(word_dist.keys())
    for i in range(0,k):
     ans.append(li[i])
    return ans
            
```
